```python
import numpy as amir
sig = lambda x: 1 / (1 + amir.exp(-x))
x, y = amir.array([[0, 0], [0, 1], [1, 0], [1, 1]]), amir.array([[0.85], [0], [0], [1]])  
w1 = amir.random.randn(2, 4)
w2= amir.random.randn(4, 1)
b1 = amir.random.randn(1, 4)
b2= amir.random.randn(1, 1)
for i in range(5000):
    mn = sig(x @ w1 + b1)
    am = sig( mn@ w2 + b2)  

print(am)  
```
